```python
import pandas as pd

df = pd.read_csv(
    '../data/feed-views.log',
    sep = '\t',
    skiprows=[2, 3],
    skipfooter=2,
    names=["datetime", "user"],
    index_col="datetime",
    engine="python"
)
```

```python
df.index.name = "date_time"
```

```python
df.to_csv(
    "../data/feed-views-semicolon.log",
    sep=";"
)
```
