```python
fines = concat_rows.copy()
```

```python
new_fines_rows = pd.DataFrame({
    'CarNumber': ['AAA111', 'BBB222', 'CCC333', 'DDD444', 'EEE555'],
    'Refund': [1, 2, 1, 2, 1],
    'Fines': [1500, 2200, 3100, 1800, 2600],
    'Make': ['Ford', 'Toyota', 'BMW', 'Ford', 'Toyota'],
    'Model': ['Focus', 'Camry', 'X5', 'Focus', 'Corolla'],
    'Year': [2005, 2006, 2007, 2008, 2009]
})
```

```python
fines = pd.concat([fines, new_fines_rows], ignore_index=True)
len(fines)
```

```python
owners = owners.iloc[:-20]
len(owners)
```

```python
new_owner_rows = pd.DataFrame({
    'CarNumber': ['ZZZ999', 'YYY888', 'XXX777'],
    'SURNAME': ['SMITH', 'JOHNSON', 'BROWN']
})

owners = pd.concat([owners, new_owner_rows], ignore_index=True)
```

```python
set(fines['CarNumber']) - set(owners['CarNumber'])
```

```python
set(owners['CarNumber']) - set(fines['CarNumber'])
```

```python
inner_join = fines.merge(
    owners,
    on='CarNumber',
    how='inner'
)
```

```python
len(inner_join)
```

```python
left_join = fines.merge(
    owners,
    on='CarNumber',
    how='left'
)
len(left_join)
```

```python
right_join = fines.merge(
    owners,
    on='CarNumber',
    how='right'
)
len(right_join)
```

```python
outer_join = fines.merge(
    owners,
    on='CarNumber',
    how='outer'
)
len(outer_join)
```

```python
print(len(fines),len(owners),len(inner_join),len(left_join),len(right_join),len(outer_join))
```
