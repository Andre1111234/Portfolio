```python
import pandas as pd

df = pd.read_csv(
    '../data/auto.csv',
    sep = ','
)
```

```python
df.count()
```

```python
df = df.set_index('ID')
```

```python
len(df)
```

```python
df = df.drop_duplicates(
    subset=['CarNumber', 'Make_n_model', 'Fines'],
    keep='last'
)
```

```python
len(df)
```

```python
df.count()
```

```python
df.isna().sum()
```

```python
df = df.dropna(thresh=len(df) - 500, axis=1)
```

```python
df.isna().sum()
```

```python
df['Refund'] = df['Refund'].ffill()
```

```python
df.isna().sum()
```

```python
mean_fines = df['Fines'].mean()
df['Fines'] = df['Fines'].fillna(mean_fines)
```

```python
df[['Make', 'Model']] = df['Make_n_model'].apply(
    lambda x: pd.Series(x.split(' ', 1))
)
```

```python
df = df.drop(columns=['Make_n_model'])
```

```python
df.head()
```

```python
df.to_json('../data/auto.json', orient='records')
```
