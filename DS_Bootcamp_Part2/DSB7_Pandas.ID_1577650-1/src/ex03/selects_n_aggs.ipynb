```python
violations = df.groupby(df.index)['Fines'].count().sort_values(ascending=False)
```

```python
top_number = violations.idxmax()
df.loc[top_number]
```

```python
total_paid = df.groupby(df.index)['Fines'].sum().sort_values(ascending=False)
```

```python
top_sum = total_paid.idxmax()
df.loc[top_sum]
```
