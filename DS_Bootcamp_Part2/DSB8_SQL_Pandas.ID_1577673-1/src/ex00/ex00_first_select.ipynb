```python
import pandas as pd
import sqlite3

con = sqlite3.connect('../data/checking-logs.sqlite')
```

```python
schema = pd.read_sql_query("PRAGMA table_info(pageviews);", con)

print(schema)
```

```python
f_10 = pd.read_sql_query('SELECT * FROM pageviews LIMIT 10',con)
print(f_10)
```

```python
pageviews = pd.read_sql_query("""
    SELECT uid, datetime
    FROM pageviews
    WHERE uid LIKE 'user_%'
    ORDER BY uid ASC
""", con)
```

```python
pageviews['datetime'] = pd.to_datetime(pageviews['datetime'])
pageviews = pageviews.set_index('datetime')
```
