# Киноисследование: Анализ зрительских предпочтений

## Обзор проекта
Анализ данных MovieLens для понимания:
- Паттернов оценок фильмов
- Временных тенденций
- Популярности жанров и фильмов
- Поведения пользователей


```python
from movielens_analysis import Ratings, Tags, Links

ratings = Ratings("ratings.csv", "movis.csv")
movies = ratings.movies
users = ratings.user
tags = Tags("tags.csv")
links = Links("links.csv")

print("Загрузка данных...")
%timeit Ratings("ratings.csv", "movies.csv")
%timeit ratings.movies
%timeit ratings.user
%timeit Tags("tags.csv")
%timeit Links("links.csv")
```

## Часть 1: Анализ рейтингов (Ratings & Movies)

### Временной анализ

```python
print("Распределение оценок по годам ")
years_dist = ratings.dist_by_year()
%timeit ratings.dist_by_year()
years_dist
```

```python
print("\nГоды выпуска фильмов")
release_dist = movies.dist_by_release()
%timeit movies.dist_by_release()
release_dist
```

### Распределение оценок

```python
print("Распределение оценок пользователей")
rating_dist = ratings.dist_by_rating()
%timeit ratings.dist_by_rating()
rating_dist
```

### Топ фильмов

```python
print("Сравнение рейтингов фильмов\n")

print("Топ-5 по количеству оценок:")
top_by_count = ratings.top_by_num_of_ratings(5)
%timeit ratings.top_by_num_of_ratings(5)
top_by_count
```

```python
print("\nТоп-5 по среднему рейтингу:")
top_by_avg = movies.top_by_ratings(5, 'average')
%timeit movies.top_by_ratings(5, 'average')
top_by_avg
```

### Анализ жанров

```python
print("Популярность жанров")
genre_dist = movies.dist_by_genres()
%timeit movies.dist_by_genres()
genre_dist
```

```python
print("Фильмы с наибольшим количеством жанров\n")
multi_genre = movies.most_genres(5)
print("Топ-5 фильма с множеством жанров:")
%timeit movies.most_genres(5)
multi_genre
```

### Самые спорные фильмы

```python
print("Самые спорные фильмы\n")
controversial = movies.top_controversial(5)
print("Фильмы с наибольшей дисперсией оценок:")
%timeit movies.top_controversial(5)
controversial
```

## Часть 2: Анализ пользователей (Users)

### Активность пользователей

```python
print("Активность пользователей")
user_activity = users.top_by_num_of_user()
%timeit users.top_by_num_of_user()
user_activity
```

### Паттерны оценок пользователей

```python
print("Сравнение оценок пользователей\n")

print("Средние оценки пользователей:")
avg_ratings = users.top_by_ratings_user('average')
%timeit users.top_by_ratings_user('average')
avg_ratings
```

```python
print("\nМедианные оценки пользователей:")
median_ratings = users.top_by_ratings_user('median')
%timeit users.top_by_ratings_user('median')
median_ratings
```

### Непредсказуемые пользователи

```python
print("Самые непредсказуемые пользователи\n")
user_variance = users.top_controversial_user(5)
print("Пользователи с наибольшей дисперсией оценок:")
%timeit users.top_controversial_user(5)
user_variance
```

## Часть 3: Анализ тегов (Tags)

### Популярные теги

```python
print("Популярность тегов\n")
popular_tags = tags.most_popular(5)
print("Самые популярные теги:")
%timeit tags.most_popular(5)
popular_tags
```

```python
print("Теги по категориям")
comedy_tags = tags.tags_with('comedy')
print("Теги содержащие 'comedy':")
%timeit tags.tags_with('comedy')
comedy_tags
```

### Структура тегов

```python
print("Структура тегов\n")

print("Теги с наибольшим количеством слов:")
wordy_tags = tags.most_words(5)
%timeit tags.most_words(5)
wordy_tags
```

```python
print("Самые длинные теги (по символам):")
longest_tags = tags.longest(5)
%timeit tags.longest(5)
longest_tags
```

```python
print("Пересечение этих характеристик:")
most_w_l = tags.most_words_and_longest(5)
%timeit tags.most_words_and_longest(5)
most_w_l
```
